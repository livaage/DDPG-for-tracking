```python
def find_nearest(particle, hit_id): 
        hit = particle[particle['hit_id']==hit_id]
        particle = particle[particle['hit_id']!=hit_id]
        distances = [] 
        #it's a big search, converting to list from pandas save an order of magnitude in time 
        rlist = particle.r.tolist()
        zlist = particle.z.tolist() 
        particle = particle[particle['r'] > hit.r.values[0]]
        for ix in range(len(particle)): 
            row = particle.iloc[ix, ]
            distance = np.sqrt((hit.r - rlist[ix])**2 + (hit.z-zlist[ix])**2)
            distances.append(distance)
        if len(distances) ==0: 
            return (0, 0)   
        where = np.argmin(distances)
        return(np.array([hit.hit_id.values[0], particle.iloc[where,:].hit_id]))
```
