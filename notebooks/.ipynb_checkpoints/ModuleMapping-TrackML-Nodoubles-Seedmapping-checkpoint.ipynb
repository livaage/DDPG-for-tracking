```python
def read_hits(filenumber): 
    prefix = '/home/lhv14/exatrkx/Tracking-ML-Exa.TrkX/alldata/train_2/event00000'+str(2820+filenumber)


    hits, particles, truth = trackml.dataset.load_event(
            prefix, parts=['hits', 'particles', 'truth'])


    hits['r'] = np.sqrt(hits.x**2 + hits.y**2)/10

    particles['pt'] = np.sqrt(particles.px**2 + particles.py**2)

    truth = (truth[['hit_id', 'particle_id']]
                 .merge(particles[['particle_id', 'pt', 'nhits']], on='particle_id'))

    phi = np.arctan2(hits.y, hits.x)

    # Select the data columns we need
    hits = (hits[['hit_id', 'x', 'y', 'z', 'r', 'layer_id', 'volume_id', 'module_id']]
            .assign(phi=phi)
            .merge(truth[['hit_id', 'particle_id', 'pt', 'nhits']], on='hit_id'))

    # Remove duplicate hits

    hits['z'] = hits['z']/10
    hits['x'] = hits['x']/10
    hits['y'] = hits['y']/10
    

    hits = hits.sort_values(['r', 'z'])

    print("lne0", len(hits))
    #hits['new_module_id'] = hits['volume_id']*100000000+ hits['layer_id']*100000 + hits['module_id']
    mods = hits.set_index(['volume_id', 'layer_id', 'module_id']).index

    #hits['discrete_module_id'] = [remap_modules_dic[val] for val in mods]
    
    c = hits.groupby(['particle_id']).count()
    
    hits = hits[hits['particle_id'].isin(c[c['hit_id'] > 6].index)]
    
    print("1", len(hits[hits['particle_id']==allowed[0]]))
    lids = pd.DataFrame(unique_layer_ids).reset_index().melt(id_vars=['index'], value_vars=[8, 13, 17, 7, 9, 12, 14, 16, 18])
    lids = lids.rename({'index':'layer_id', 'variable':'volume_id', 'value':'unique_layer_id'}, axis=1)
    lids[["volume_id", "layer_id"]] = lids[["volume_id", "layer_id"]].apply(pd.to_numeric)

    hits = hits.merge(lids, on=['volume_id', 'layer_id'], how='left')
    hits = hits.sort_values(['r', 'z'])
    print("2",  len(hits[hits['particle_id']==allowed[0]]))

    
    smallest_rs = hits.groupby(['particle_id', 'unique_layer_id'])['r'].min()
    hits = hits.merge(smallest_rs)
    print("3",  len(hits[hits['particle_id']==allowed[0]]))

    #hits = hits.drop_duplicates(subset=['hit_id'])
    hits = hits.drop_duplicates()
    print("4",  len(hits[hits['particle_id']==allowed[0]]))

    hits_barrel = hits[hits['volume_id'].isin([8, 13, 17])]

    hits_barrel['new_module_id'] = [str(int(row[0])) + str(int(row[1])) + str(int(round(row[2], 0))) for row in hits_barrel[['volume_id', 'unique_layer_id','z']].values] 

    hits_endcaps = hits[~hits['volume_id'].isin([8, 13, 17])]
    hits_endcaps['new_module_id'] = [str(int(row[0])) + str(int(row[1])) + str(int(round(row[2], 0))) for row in hits_endcaps[['volume_id', 'unique_layer_id','r']].values] 

    hits = pd.concat([hits_barrel, hits_endcaps])

    print("5",  hits[hits['particle_id']==allowed[0]])

    hits[['next_mod_id', 'connecting_layer_id']] = hits.groupby('particle_id')['new_module_id', 'unique_layer_id'].shift(-1)

    hits = hits.merge(allowed_connections, on=['unique_layer_id', 'connecting_layer_id'], how='inner')
    hits = hits.sort_values(['r', 'z'])

    #hits = hits.dropna()
    hits = hits[hits['unique_layer_id']!=hits['connecting_layer_id']]
    print(6, hits[hits['particle_id']==allowed[0]])

    
    return hits  
```

```python
class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        if isinstance(obj, np.floating):
            return float(obj)
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        return super(NpEncoder, self).default(obj)
```
